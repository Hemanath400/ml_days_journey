```python
#GOAL: Understand that EVERY data point is a vector
"""
Concepts:
- What is a vector? (magnitude + direction)
- Vector representation in ML: [feature1, feature2, ...]
- Your customer = vector [tenure, monthly_charges, age]
"""
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
customer1=np.array([12,85.5,35]) #(tenure,Monthly_charges,age)
customer2=np.array([24,65.0,42])
customer3=np.array([6,120.5,28])

print(f'customer1(short term):{customer1}')
print(f'customer2(longterm):{customer2}')
print(f'customer3(new and high spend):{customer3}')
```

```python
plt.Figure(figsize=(10,6))
plt.scatter([customer1[0],customer2[0],customer3[0]],
            [customer1[1],customer2[1],customer3[1]],
            s=200,c=['red','blue','green'])
plt.xlabel('tenure(months)',fontsize=12)
plt.ylabel('Monthly charges($)',fontsize=12)
plt.title('Customer as Vectors in 2d space',fontsize=14)
for i,txt in enumerate(['C1','C2','C3']):
    plt.annotate(txt,([customer1[0],customer2[0],customer3[0]][i],
                      [customer2[1],customer2[1],customer2[1]][i]),
                      fontsize=12,fontweight='bold')
    
plt.grid(True,alpha=0.3)
plt.savefig('day1_vector.png')
plt.show()
```
