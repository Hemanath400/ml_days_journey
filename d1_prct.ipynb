```python
import numpy as np

# Original vectors from the lesson
customer1 = np.array([12, 85.5, 35])
customer2 = np.array([24, 65.0, 42])
customer3 = np.array([6, 120.5, 28])
myself = np.array([8, 79.99, 32])

# Calculate magnitude using Euclidean distance formula: ‚àö(x‚ÇÅ¬≤ + x‚ÇÇ¬≤ + x‚ÇÉ¬≤)
def vector_magnitude(vector):
    return np.sqrt(np.sum(vector**2))

# Calculate magnitudes
mag1 = vector_magnitude(customer1)
mag2 = vector_magnitude(customer2)
mag3 = vector_magnitude(customer3)
mag_me = vector_magnitude(myself)

print("üìä Vector Magnitudes:")
print(f"Customer 1 magnitude: {mag1:.2f}")
print(f"Customer 2 magnitude: {mag2:.2f}")
print(f"Customer 3 magnitude: {mag3:.2f}")
print(f"My magnitude: {mag_me:.2f}")

# Using numpy's built-in norm function (easier way)
mag1_np = np.linalg.norm(customer1)
print(f"\nUsing numpy.linalg.norm: {mag1_np:.2f} (same result!)")

# Understanding what magnitude represents
print(f"\nüí° Interpretation: Magnitude represents the 'total size' of the customer")
print(f"   Customer 1 total magnitude: {mag1:.2f} (combination of tenure, charges, age)")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate Euclidean distances between customers
def euclidean_distance(v1, v2):
    return np.sqrt(np.sum((v1 - v2)**2))

# Compare in full 3D space
dist_1_2 = euclidean_distance(customer1, customer2)
dist_1_3 = euclidean_distance(customer1, customer3)
dist_2_3 = euclidean_distance(customer2, customer3)
dist_me_1 = euclidean_distance(myself, customer1)
dist_me_2 = euclidean_distance(myself, customer2)
dist_me_3 = euclidean_distance(myself, customer3)

print("üîç Customer Similarity (smaller distance = more similar):")
print(f"Distance between Customer 1 and 2: {dist_1_2:.2f}")
print(f"Distance between Customer 1 and 3: {dist_1_3:.2f}")
print(f"Distance between Customer 2 and 3: {dist_2_3:.2f}")
print(f"\nDistance between Me and Customer 1: {dist_me_1:.2f}")
print(f"Distance between Me and Customer 2: {dist_me_2:.2f}")
print(f"Distance between Me and Customer 3: {dist_me_3:.2f}")

# Visual analysis
print("\nüëÅÔ∏è Visual Analysis (from the scatter plot):")
print("- Customer 1 (red) and Customer 2 (blue) appear relatively close in tenure-charge space")
print("- Customer 3 (green) is far from both due to high monthly charges but low tenure")
print("- Most similar pair visually: Customers 1 and 2 (both moderate tenure and charges)")
print("- Most different: Customer 3 stands alone with high charges, low tenure")

# Which customer am I most similar to?
most_similar = np.argmin([dist_me_1, dist_me_2, dist_me_3]) + 1
print(f"\n‚úÖ Based on distances, I'm most similar to Customer {most_similar}")

# Create enhanced visualization
plt.figure(figsize=(12, 8))

# Plot all customers
customers = [customer1, customer2, customer3, myself]
colors = ['red', 'blue', 'green', 'purple']
labels = ['C1', 'C2', 'C3', 'Me']

for i, (c, color, label) in enumerate(zip(customers, colors, labels)):
    plt.scatter(c[0], c[1], s=200, c=color, label=f'{label} ({c[0]}, {c[1]})')
    plt.annotate(label, (c[0], c[1]), fontsize=12, fontweight='bold', 
                xytext=(5, 5), textcoords='offset points')

plt.xlabel('Tenure (months)', fontsize=12)
plt.ylabel('Monthly Charges ($)', fontsize=12)
plt.title('Customer Vectors: Who is Most Similar?', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('day1_similarity.png')
plt.show()
```
